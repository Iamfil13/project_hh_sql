Среди регионов, в которых нет вакансий, найдите тот, в котором наибольшее количество работодателей.


```python
# текст запроса
# Регион без вакансий с наибольшим количеством работодателей
query_5_2 = """
WITH employers_cnt AS (
    SELECT area, COUNT(*) AS employers_cnt
    FROM public.employers
    GROUP BY area
),
vacancies_cnt AS (
    SELECT area_id, COUNT(*) AS vacancies_cnt
    FROM public.vacancies
    GROUP BY area_id
)
SELECT
    a.name AS area_name,
    COALESCE(e.employers_cnt, 0) AS employers_cnt,
    COALESCE(v.vacancies_cnt, 0) AS vacancies_cnt
FROM public.areas a
LEFT JOIN employers_cnt e ON e.area = a.id
LEFT JOIN vacancies_cnt v ON v.area_id = a.id
WHERE COALESCE(v.vacancies_cnt, 0) = 0
ORDER BY employers_cnt DESC
LIMIT 1;
"""
```

```python
# результат запроса
df = pd.read_sql(query_5_2, connection)
print(df.iloc[0])
```

7. Для компании «Яндекс» выведите список регионов-миллионников, в которых представлены вакансии компании, вместе с количеством вакансий в этих регионах. Также добавьте строку Total с общим количеством вакансий компании. Результат отсортируйте по возрастанию количества.

Список городов-милионников надо взять [отсюда](https://ru.wikipedia.org/wiki/%D0%93%D0%BE%D1%80%D0%BE%D0%B4%D0%B0-%D0%BC%D0%B8%D0%BB%D0%BB%D0%B8%D0%BE%D0%BD%D0%B5%D1%80%D1%8B_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8). 

Если возникнут трудности с этим задание посмотрите материалы модуля  PYTHON-17. Как получать данные из веб-источников и API. 

```python
# текст запроса
# Список миллионников
million_cities_list = [
    'Москва', 'Санкт-Петербург', 'Новосибирск', 'Екатеринбург', 'Казань',
    'Нижний Новгород', 'Челябинск', 'Красноярск', 'Самара', 'Уфа',
    'Ростов-на-Дону', 'Омск', 'Воронеж', 'Пермь', 'Волгоград', 'Краснодар'
]

# Динамически формируем IN-условие
placeholders = ','.join([f"'{city}'" for city in million_cities_list])

query_5_7 = f"""
WITH million_cities AS (
  SELECT id FROM areas 
  WHERE name IN ({placeholders})
),
yandex_vac_in_million AS (
  SELECT COUNT(*) AS total_cnt
  FROM vacancies v
  JOIN employers e ON v.employer_id = e.id
  JOIN areas a ON v.area_id = a.id
  WHERE e.name = 'Яндекс' AND a.name IN ({placeholders})
)
SELECT a.name, COALESCE(y.cnt, 0)::int AS cnt
FROM million_cities m
JOIN areas a ON m.id = a.id
LEFT JOIN (
  SELECT v.area_id, COUNT(*) AS cnt
  FROM vacancies v
  JOIN employers e ON v.employer_id = e.id
  WHERE e.name = 'Яндекс'
  GROUP BY v.area_id
) y ON a.id = y.area_id

UNION ALL

SELECT 'Total' AS name, yvm.total_cnt::int AS cnt
FROM yandex_vac_in_million yvm

ORDER BY cnt DESC NULLS LAST, name
"""
```

```python
# результат запроса
df = pd.read_sql_query(query_5_7, connection)
print(df.iloc[:, [0, 1]])
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в **среднем** указывают для каждого типа требуемого опыта (уникальное значение из поля *experience*). 

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа. 

```python
query_6_6 = """
SELECT
    ROUND(
        AVG(
            CASE 
                WHEN salary_from IS NOT NULL AND salary_to IS NOT NULL 
                    THEN (salary_from + salary_to) / 2.0
                WHEN salary_from IS NOT NULL THEN salary_from
                WHEN salary_to IS NOT NULL THEN salary_to
            END
        )::numeric, 
        0
    ) AS avg_salary_ds_3_6
FROM vacancies
WHERE
    (
        LOWER(name) LIKE '%data scientist%' OR
        LOWER(name) LIKE '%data science%' OR
        LOWER(name) LIKE '%исследователь данных%' OR
        (name LIKE '%ML%' AND LOWER(name) NOT LIKE '%html%') OR
        LOWER(name) LIKE '%machine learning%' OR
        LOWER(name) LIKE '%машинн%обучен%'
    )
    AND NOT (
        LOWER(name) LIKE '%junior%' OR 
        experience = 'Нет опыта'
    )
    AND experience = 'От 3 до 6 лет'
    AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)
"""
```

```python
# результат запроса
df = pd.read_sql_query(query_6_6, connection)
print(df.iloc[0, 0])
```

- Большинство работодателей публикуют **1–2 вакансии**, рынок сильно фрагментирован  
- Крупные компании (Яндекс, Сбер, VK и др.) формируют значимую долю предложений, но **географически сильно ограничены**

### Вывод  
Рынок труда представлен большим количеством мелких и средних игроков, а «гиганты» дают масштаб, но не разнообразие по локациям.

---

## 2. Data Science — ниша с премиальным спросом

### Объём рынка DS
- Всего вакансий, связанных с Data Science: **~1770**
- После исключения junior / стажировок / «без опыта»: **~660–700 вакансий**
- Доля DS-вакансий: **около 3%** от всего рынка

**DS — нишевая, но стратегически важная область**

---

## 3. DS-вакансии по уровню опыта

| Опыт        | Кол-во | Доля | Средняя ЗП |
|------------|-------:|-----:|-----------:|
| 1–3 года   | ~330   | ~50% | ~200 тыс. ₽ |
| 3–6 лет    | ~250   | ~38% | **243 000 ₽** |

### Интерпретация
- **Пик спроса — специалисты с опытом 1–3 года**  
  → рынок активно растёт и нуждается в middle-уровне  
- **Junior-позиций очень мало**  
  → высокая конкуренция, дорогое обучение, риски для бизнеса  
- Зарплаты 3–6 лет — **премиальный сегмент**, в 2–3 раза выше средней по рынку

### Почему так?
- DS требует сразу набора сложных навыков (Python, SQL, ML)
- Junior редко приносят быструю бизнес-ценность
- Компании предпочитают «готовых» специалистов

**Прогноз:** рост DS-рынка на **20–30% к 2027 году** за счёт AI / LLM / автоматизации.

---

## 4. Навыки Data Scientist

### Частота ключевых навыков
- **Python** — ~90% вакансий  
- **SQL / Postgres** — ~70–80% вакансий  

Python + SQL — **обязательный входной билет** в профессию

### Среднее количество навыков
- В среднем в одной DS-вакансии указывают **~6–6.5 ключевых навыков**
- Навыки разделены **символом табуляции (CHR(9))** — это артефакт парсинга, а не особенность рынка

### Вывод
DS-вакансии отличаются:
- чёткими требованиями,
- высокой насыщенностью по навыкам,
- низкой толерантностью к «размытым» профилям.

## 5. География рынка труда

| Город        | Кол-во вакансий |
|-------------|----------------:|
| Москва      | ~5300 |
| Санкт-Петербург | ~2850 |
| Новосибирск / Минск / Алматы | ~500 |
| Остальные города | значительно меньше |

### Интерпретация
- **70%+ вакансий сконцентрированы в Москве и СПб**
- Даже крупные компании:
  - Яндекс — в основном Москва и несколько крупных регионов
- Регионы без вакансий:
  - либо ориентированы на офлайн-рынок,
  - либо данные заполнены неполно

### Вывод
Рынок труда **сильно централизован**.  
Удалёнка частично снижает эффект, но не устраняет его.
